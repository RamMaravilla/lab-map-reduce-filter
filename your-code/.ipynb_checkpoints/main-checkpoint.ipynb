# Challenge 1 - Mapping

#### We will use the map function to clean up words in a book.

In the following cell, we will read a text file containing the book The Prophet by Khalil Gibran.

```python
# Run this code:

location = '../58585-0.txt'
with open(location, 'r', encoding="utf8") as f:
    prophet = f.read().split(' ')    #ESTO LEE PROFETA Y LO PARTE EN PALABRAS
```

#### Let's remove the first 568 words since they contain information about the book but are not part of the book itself. 

Do this by removing from `prophet` elements 0 through 567 of the list (you can also do this by keeping elements 568 through the last element).

```python
# Your code here:

#Si queremos quitar palabras tenemos que:
    #1) Crear un index para saber qué palabras son las indicadas
    #2) Usar la palabra filtro

profeta_limpio = list(filter(lambda x: prophet.index(x) > 567, prophet))  #Se tiene que poner en lista para que filter funcione
                                #Se pone filter para que filtre
                                #Lambda sustitutye a x

profeta_limpio
```

If you look through the words, you will find that many words have a reference attached to them. For example, let's look at words 1 through 10.

```python
# Your code here:
profeta_limpio[1:10]
```

#### The next step is to create a function that will remove references. 

We will do this by splitting the string on the `{` character and keeping only the part before this character. Write your function below.

```python
import re
def reference(x):
    '''
    Input: A string
    Output: The string with references removed
    
    Example:
    Input: 'the{7}'
    Output: 'the'
    '''
    #Tenemos que usar re(regular expression) y replace. re.sub

    return re.sub('\{.+\}','',x)   #Hay que hacer un regrex. El regex se pone entre comillas  
                            # \ busca al símbolo inmediatamente a continuación "{"
                            # '{' es el que se tiene que buscar
                            # '.' busca a cualquier caracter excepto \n (line terminator)
                            # '+' le dice que busque todo lo que está inmediataemente antes muchas veces
                            # se reemplaza todo lo anterior por nada ('') 
    # Your code here:
       #Tenemos que usar re(regular expression) y replace. re.sub
    
print(reference(profeta_limpio[1]))
```

Now that we have our function, use the `map()` function to apply this function to our book, The Prophet. Return the resulting list to a new list called `prophet_reference`

```python
# Your code here:
prophet_reference = list(map(reference, profeta_limpio))
prophet_reference
```

Another thing you may have noticed is that some words contain a line break. Let's write a function to split those words. Our function will return the string split on the character `\n`. Write your function in the cell below.

```python
def line_break(x):
    '''
    Input: A string
    Output: A list of strings split on the line break (\n) character
        
    Example:
    Input: 'the\nbeloved'
    Output: ['the', 'beloved']
    '''
    
    # Your code here:
    #return re.sub('\n', " ", x)   #esto no funciona porque no tengo idea #$%&!
    return x.split("\n")           #esto ya lo toma directamente y lo separa. No se sube desde prophet_reference
                                #se tiene que aplicar desde la siguiente lista
    
#print(line_break(prophet_reference[3]))  #TE REGRESA NONE
print(line_break('to\nreturn'))
```

Apply the `line_break` function to the `prophet_reference` list. Name the new list `prophet_line`.

```python
# Your code here:
prophet_line = list(map(line_break, prophet_reference))  #Lo hace una lista de listas....ok
prophet_line
```

If you look at the elements of `prophet_line`, you will see that the function returned lists and not strings. Our list is now a list of lists. Flatten the list using list comprehension. Assign this new list to `prophet_flat`.

```python
# Your code here:

prophet_flat = ' '.join([l[0] for l in prophet_line]).split()  #GRACIAS STOCKERFLOW! Esto es para lista de listas con una sola palabra
                    #list2 = ' '.join([' '.join(strings) for strings in list1]).split() #Si las listas tienen más de una palabra

prophet_flat
```

We will start by writing a function that takes two strings and concatenates them together with a space between the two strings.

```python
def concat_space(a, b):
    '''
    Input:Two strings
    Output: A single string separated by a space
        
    Example:
    Input: 'John', 'Smith'
    Output: 'John Smith'
    '''

    # Your code here:
    #return "".join(a,b)  
    #return % (a,b)
    return a+" "+b

print(concat_space("No","quiero"))
```
